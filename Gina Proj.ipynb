```python
#Dependencies
import pandas as pd
import scipy.stats as st
import numpy as np

#Data files
#Meter paths do not change ever (ie. DO NOT TOUCH)
FUU_Meters_path = "Resources/FtUnionUpper_Meters.csv"
FUL_Meters_path = "Resources/FtUnionLower_Meters.csv"

#These change by month
FU_Scada_path = "Feb_21/2021-02_FtUnion_SCADA.csv"
FU_PRAMS_path = "Feb_21/2021-02_FtUnion_PRAMS_dailies_rpt.csv"

#This will output each time. Rename in documents so the data is saved
#May have to delete from local after each run else error, but maybe not
Meter_SCADA_Well_Name_file = "outputs/Meter_SCADA_Well_Name_file.csv"
SCADA_PRAMS_combo_file="outputs/SCADA_PRAMS_combo_file.csv"

#Read the data
FUU_Meters = pd.read_csv(FUU_Meters_path)
FUL_Meters = pd.read_csv(FUL_Meters_path)
FU_Scada = pd.read_csv(FU_Scada_path)
FU_PRAMS = pd.read_csv(FU_PRAMS_path)

FUU_Meters.head()
```

```python
FUL_Meters.head()
```

```python
FU_Scada.head()
```

```python
#Combine meter dicts into single dataset (Lower added to end of upper)
FU_combo = pd.concat([FUU_Meters, FUL_Meters])
FU_combo.head()
```

```python
#Combine combo meters with Scada
FU_analysis = pd.merge(FU_combo, FU_Scada, on="MeterId")

#Display merge
FU_analysis.head()
```

```python
FU_relevant_data = FU_analysis[['Date','name','Product_Name','MeterId','Site_Name','VolYdy']]
FU_relevant_data.head()
```

```python
FU_SCADA_clean = FU_relevant_data.rename(columns={"Date":"Date","name":"Route","Product_Name":"Type","MeterId":"MeterID",
                                            "Site_Name":"Well","VolYdy":"SCADA Volumes"})
FU_SCADA_clean
```

```python
FU_SCADA_clean.to_csv(Meter_SCADA_Well_Name_file)
```
